```python
import matplotlib.pyplot as plt
import pandas as pd

vgg_history = pd.read_csv(r'../logs/history_vgg_2009_150_64.csv')
inception_history = pd.read_csv(r'../logs/history_inception_2009_150_64.csv')
resnet_history = pd.read_csv(r'../logs/history_resnet_2009_150_64.csv')
```

```python
train_colors = ['crimson', 'mediumseagreen', 'royalblue']
val_colors = ['lightcoral', 'lightgreen', 'lightskyblue']
```

```python
for i, history in enumerate([vgg_history, inception_history, resnet_history]):
    plt.figure(figsize=(5, 5))
    # plt.ylim([0.075, 0.275])
    # plt.xlim([0, 51])
    plt.plot(range(1, 51), history['loss'], label='Zbir treningowy', color=train_colors[i])
    plt.plot(range(1, 51), history['val_loss'], label='Zbir walidacyjny', color=val_colors[i])
    # plt.grid()
    plt.legend(loc='upper right')
    # plt.title('Model loss')
    plt.ylabel('Strata kontrastowa')
    plt.xlabel('Epoka')
    plt.show()

    plt.figure(figsize=(5, 5))
    # plt.ylim([0, 1])
    # plt.xlim([0, 51])
    plt.plot(range(1, 51), history['accuracy'], label='Zbir treningowy', color=train_colors[i])
    plt.plot(range(1, 51), history['val_accuracy'], label='Zbir walidacyjny', color=val_colors[i])
    # plt.grid()
    plt.legend(loc='lower right')
    # plt.title('Model accuracy')
    plt.ylabel('Dokadno [%]')
    plt.xlabel('Epoka')
    plt.show()
```
